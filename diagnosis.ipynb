### Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

### Loading dataset

```python
# Load Breast Cancer dataset
data = load_breast_cancer()

X = pd.DataFrame(data.data, columns=data.feature_names) # feature
y = data.target # target

print("Dataset Shape:", X.shape)
print("\nTarget Classes:")
print(pd.Series(y).value_counts())
```

### Splitting data

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2, # 20% of the data
    random_state=42
)

print("Training Set:", X_train.shape) # size of the training set
print("Testing Set:", X_test.shape) # size of the testing set
```

### Scaling features

```python
# feature standardization
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### model training

```python
model = LogisticRegression(max_iter=5000)

model.fit(X_train_scaled, y_train) # model training

print("Model trained successfully!")
```

### Prediction

```python
y_pred = model.predict(X_test_scaled)

print("Predictions:")
print(y_pred[:10]) # display first 10 predictions
```

### Model evaluation

```python
# calculating accuracy score
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy score:\n{round(accuracy * 100, 2)}%') # rounding off to 2dp
```

### Classification report

```python
classification_report(y_test, y_pred)
```

### Confusion Matrix Visualization

```python
confusion_mat = confusion_matrix(y_test, y_pred) # compares actual diagnoses with predicted ones
# y_pred: these are model predicted diagnoses
# y_test: these are the actual diagnoses

plt.figure(figsize = (10, 5))
sns.heatmap(confusion_mat ,annot = True, fmt = 'd')

plt.title('Disease Diagnosis Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')

plt.show()
```

### Simulation

```python
# Simulating a new patient
patient = X.iloc[[0]]
patient_scaled = scaler.transform(patient)
prediction = model.predict(patient_scaled) # predicting diagnosis

if prediction[0] == 1:
    print("Diagnosis Results: Benign (No Cancer Detected)")
else:
    print("Diagnosis Results: Malignant (Cancer Detected)")
```
